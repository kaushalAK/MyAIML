```python
first_frame = None
status_list = [None,None]
times = []

df = pd.DataFrame(columns=["Start","End"])

face_cascade = cv2.CascadeClassifier('FaceDetect.xml')

video = cv2.VideoCapture(0)

val = 1

while True:
     
    check , frame = video.read()
    status = 0
    gray = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    gray = cv2.GaussianBlur(frame,(21,21),0)
    
    if first_frame is None:
        first_frame = gray
        continue
        
#difference in the frames

    #delta_frame = cv2.absdiff(first_frame,gray)
    #thresh_data = cv2.threshold(delta_frame, 30, 255, cv2.THRESH_BINARY)[1]
    
    #thresh_data = cv2.dilate(thresh_data,None,iterations =0)
    
    #(_,cnts,_) = cv2.findContours(cv2.cvtColor(thresh_data.copy(), cv2.COLOR_RGB2GRAY),cv2.RETR_EXTERNAL ,cv2.CHAIN_APPROX_SIMPLE)
   


    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(
        gray,     
        scaleFactor=1.2,
        minNeighbors=5,     
        minSize=(20, 20)
    )
    for (x,y,w,h) in faces:
        
        frame = cv2.rectangle(frame,(x,y),(x+w,y+h),(0,255,0),3)
        #roi_gray = gray[y:y+h, x:x+w]
        #roi_color = frame[y:y+h, x:x+w]  
        
        print(frame)
    cv2.imshow('video',frame)
   
    key = cv2.waitKey(1)

    if key == ord('q'):
        break



    
video.release()
cv2.destroyAllWindows()
```
